```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_kerala_house():
    fig, ax = plt.subplots(figsize=(10, 10))

    # Plot ground floor
    ax.add_patch(patches.Rectangle((0, 0), 7, 7, fill=None, edgecolor='black', linewidth=2))  # Outer walls
    ax.add_patch(patches.Rectangle((2, 2), 3, 3, fill=None, edgecolor='brown', linewidth=1.5))  # Central courtyard

    # Rooms on ground floor
    ax.add_patch(patches.Rectangle((0, 0), 2, 3, fill=None, edgecolor='blue', linestyle='--', linewidth=1))  # Living room
    ax.add_patch(patches.Rectangle((0, 3), 2, 4, fill=None, edgecolor='green', linestyle='--', linewidth=1))  # Kitchen
    ax.add_patch(patches.Rectangle((5, 0), 2, 3, fill=None, edgecolor='purple', linestyle='--', linewidth=1))  # Bedroom
    ax.add_patch(patches.Rectangle((5, 3), 2, 4, fill=None, edgecolor='orange', linestyle='--', linewidth=1))  # Dining

    # Annotate
    ax.text(1, 1.5, 'Living Room', fontsize=10, ha='center')
    ax.text(1, 4.5, 'Kitchen', fontsize=10, ha='center')
    ax.text(6, 1.5, 'Bedroom', fontsize=10, ha='center')
    ax.text(6, 4.5, 'Dining', fontsize=10, ha='center')
    ax.text(3.5, 3.5, 'Courtyard', fontsize=12, ha='center', color='brown')

    # Add labels
    plt.title("Kerala Style House - Ground Floor Layout")
    plt.xlim(-1, 8)
    plt.ylim(-1, 8)
    plt.grid(True)
    plt.show()

draw_kerala_house()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_kerala_house_plan():
    fig, ax = plt.subplots(figsize=(12, 12))

    # Ground Floor Layout
    ax.add_patch(patches.Rectangle((0, 0), 10, 10, fill=None, edgecolor='black', linewidth=2))  # Outer boundary
    ax.add_patch(patches.Rectangle((4, 4), 2, 2, fill=None, edgecolor='green', linewidth=2))  # Tree courtyard
    ax.add_patch(patches.Rectangle((0, 0), 3, 3, fill=None, edgecolor='blue', linestyle='--', linewidth=1))  # Parking
    ax.add_patch(patches.Rectangle((3, 0), 3, 3, fill=None, edgecolor='purple', linestyle='--', linewidth=1))  # Gym
    ax.add_patch(patches.Rectangle((0, 3), 4, 2, fill=None, edgecolor='orange', linestyle='--', linewidth=1))  # Kitchen
    ax.add_patch(patches.Rectangle((0, 5), 4, 2, fill=None, edgecolor='brown', linestyle='--', linewidth=1))  # Dining
    ax.add_patch(patches.Rectangle((7, 0), 3, 3, fill=None, edgecolor='red', linestyle='--', linewidth=1))  # Bedroom 1
    ax.add_patch(patches.Rectangle((7, 3), 3, 3, fill=None, edgecolor='cyan', linestyle='--', linewidth=1))  # TV Room
    ax.add_patch(patches.Rectangle((5, 7), 5, 3, fill=None, edgecolor='pink', linestyle='--', linewidth=1))  # Watch space

    # Ground Floor Annotations
    ax.text(1.5, 1.5, 'Parking', fontsize=10, ha='center')
    ax.text(4.5, 1.5, 'Gym', fontsize=10, ha='center')
    ax.text(2, 4, 'Kitchen', fontsize=10, ha='center')
    ax.text(2, 6, 'Dining', fontsize=10, ha='center')
    ax.text(8.5, 1.5, 'Bedroom 1', fontsize=10, ha='center')
    ax.text(8.5, 4.5, 'TV Room', fontsize=10, ha='center')
    ax.text(7.5, 8.5, 'Watch Space', fontsize=10, ha='center')
    ax.text(5, 5, 'Tree Courtyard', fontsize=12, ha='center', color='green')

    # First Floor Layout
    ax.add_patch(patches.Rectangle((0, 10), 10, 10, fill=None, edgecolor='black', linewidth=2))  # Outer boundary
    ax.add_patch(patches.Rectangle((4, 14), 2, 2, fill=None, edgecolor='green', linewidth=2))  # Tree courtyard open to terrace
    ax.add_patch(patches.Rectangle((0, 10), 5, 5, fill=None, edgecolor='purple', linestyle='--', linewidth=1))  # Bedroom 2
    ax.add_patch(patches.Rectangle((5, 10), 5, 5, fill=None, edgecolor='blue', linestyle='--', linewidth=1))  # Bedroom 3
    ax.add_patch(patches.Rectangle((2, 15), 6, 5, fill=None, edgecolor='brown', linestyle='--', linewidth=1))  # Terrace with swing

    # First Floor Annotations
    ax.text(2.5, 12.5, 'Bedroom 2', fontsize=10, ha='center')
    ax.text(7.5, 12.5, 'Bedroom 3', fontsize=10, ha='center')
    ax.text(5, 17, 'Terrace & Swing', fontsize=10, ha='center')
    ax.text(5, 15, 'Tree Courtyard Open', fontsize=12, ha='center', color='green')

    # Set plot limits and labels
    plt.title("Kerala Style 2-Floor House Plan with Terrace", fontsize=14)
    ax.set_xlim(-1, 11)
    ax.set_ylim(-1, 21)
    plt.grid(True)
    plt.show()

draw_kerala_house_plan()
```
